```python
import os, numpy as np

db_file_path = "saved_db.dat"
DIMENSION = 64
ELEMENT_SIZE = np.dtype(np.float32).itemsize

num_records = os.path.getsize(db_file_path) // (DIMENSION * ELEMENT_SIZE)
vec_db = np.memmap(db_file_path, dtype=np.float32, mode='r', shape=(num_records, DIMENSION))

print(f"Loaded vector database with {len(vec_db)} entries.")

del vec_db
```
